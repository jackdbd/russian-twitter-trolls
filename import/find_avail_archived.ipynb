# Find available caches

We first need to see if any of the Twitter user pages have been cached by [Internet Archive](https://archive.org/index.php) and its [Wayback Machine API](https://archive.org/help/wayback_api.php).

We can do this by making a request to `http://archive.org/wayback/available?url=http://twitter.com/TWITTER_SCREEN_NAME_HERE`.

This will return the url and timestamp of any caches made, if they exist.

So we iterate through the list of twitter screen names, checking the Wayback API for any available caches.

```python
import requests
```

```python
initial = "http://archive.org/wayback/available"
```

```python
%%time
items = []
# iterate through list of flagged twitter screen names
with open('./data/twitter_handle_urls.csv') as f:
    for line in f:
        params = {'url': line}
        r = requests.get(initial, params=params)
        d = r.json()
        # print(d)
        items.append(d)
```

```python
%%time
# print URLs and timestamp of any available archives
for item in items:
    if 'archived_snapshots' in item:
        if 'closest' in item['archived_snapshots']:
            print(
                item['url'],
                item['archived_snapshots']['closest']['url'],
                item['archived_snapshots']['closest']['timestamp'])
```

```python
%%time
# write URL of any available archives to file
with open('./data/avail_urls.txt', 'w') as f:
    for item in items:
        if 'archived_snapshots' in item:
            if 'closest' in item['archived_snapshots']:
                f.write(item['archived_snapshots']['closest']['url'] + '\n')
```
